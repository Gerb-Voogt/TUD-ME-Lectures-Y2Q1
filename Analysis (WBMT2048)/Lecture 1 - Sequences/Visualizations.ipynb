# Lecture 1 - Sequences
This notebook is to accompany the lecture notes and will visualize some behaviour of sequences and some theorems. The first will be the following. Firstly let's start with the import and some setup stuff.


```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

The first sequence we are going to visualize is the most uninteresting. We start with a series which takes the following form:
\begin{equation}
    a_n = n^2
\end{equation}This means each term $a_n$ in the sequence will be equal the $n$-th term squared. As we will see the sequence is kinda boring as it just becomes some arbitrary large number fairly quickly. This is an example of a sequence which diverges.

```python
n = np.arange(1, 25, step=1)
a_n = n**2

plt.scatter(n, a_n, label=r'$a_n = n^2$', zorder=2, s=15)
plt.grid(zorder=1)
plt.xlabel(r'$n$')
plt.ylabel(r'$a_n$')
plt.title('Visualizing a simple diverging sequence')
plt.legend()
```

Some sequences exihibit some more interesting behaviour. Take for example the following sequence:
\begin{equation}
    a_n = \frac{n^2}{n^2 + 1}
\end{equation}
When can evaluate what happens when this becomes infinitly long using a limit like this:
\begin{align}
    \lim_{n\to\infty} a_n = \lim_{n\to\infty} \frac{n^2}{n^2 + 1} &= \lim_{n\to\infty} \frac{1}{1+\frac{1}{n^2}}\\
                                           &= 1
\end{align}
This is an exaple of a sequence which converges to 1 single value as $n$ becomes sufficiently large. This can be see visualized in the next cel.

```python
n = np.arange(1, 25, step=1)
a_n = n**2 / (n**2 + 1)

plt.scatter(n, a_n, label=r'$a_n = \frac{n^2}{n^2 + 1}$', zorder=2, s=15)
plt.grid(zorder=1)
plt.xlabel(r'$n$')
plt.ylabel(r'$a_n$')
plt.title('Visualizing a simple converging sequence')
plt.legend()
```

A theorem discussed in the lecture notes referred to as squeeze theorem states that:

\begin{equation}
    b_n \leq a_n \leq c_n
\end{equation}

Where $b_n$ and $c_n$ are sequences we define ourself to close in on $a_n$ from both sides. We can then eveluate the limit of the sequences $b_n$ and $c_n$. If these sequences eveluate to 1 single value $a_n$ must also neccesarily approach that as it is inbetween those other 2 sequences. We can more formally formulate this as follows:
\begin{equation}
    \text{if} \; \lim_{n\to\infty} b_n = \lim_{n\to\infty} c_n = L \quad \text{then} \quad \lim_{n\to\infty} a_n = L
\end{equation}

To illustrate this let's work through an example. Let $\{ a_n \}$ be a sequence given as:
\begin{equation}
    \{a_n\} = \left\{ \frac{\sin(n)}{n} \right\}
\end{equation}
The value of $\sin(n)$ will oscilate between a minimum of $-1$ and a maximum of $+1$. This gives us an upper and lower bound for the squeeze theorem:
\begin{equation}
  \frac{-1}{n} \leq \frac{\sin(n)}{n} \leq \frac{1}{n} \qquad n>0
\end{equation}
Where $b_n=\frac{-1}{n}$ and $c_n=\frac{1}{n}$. We now eveluate the limit of $b_n$ and $c_n$ as $n$ goes to infity and find that:
\begin{equation}
  \lim_{n\to\infty} \{b_n\} = \lim_{n\to\infty} \{c_n\} = 0
\end{equation}
Since $b_n$ and $c_n$ both converge to $0$ and $a_n$ is inbetween these 2 we now can say for sure that:
\begin{equation}
  \lim_{n\to\infty} \{a_n\} = \lim_{n\to\infty} \left\{ \frac{\sin(n)}{n} \right\} = 0
\end{equation}


```python
n = np.arange(1, 25, step=1)
a_n = np.sin(n)/n
b_n = -1/n
c_n = 1/n

plt.scatter(n, a_n, label=r'$\{a_n\}$', zorder=2, s=15)
plt.scatter(n, b_n, label=r'$\{b_n\}$', zorder=2, s=15)
plt.scatter(n, c_n, label=r'$\{c_n\}$', zorder=2, s=15)
plt.grid(zorder=1)
plt.xlabel(r'$n$')
plt.ylabel(r'$a_n, b_n, c_n$')
plt.title('Visual intuition for the squeeze theorem')
plt.legend()
```
